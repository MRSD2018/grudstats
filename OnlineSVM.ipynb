```python
import os
from os.path import isfile, join
from sklearn.linear_model import SGDClassifier
import matplotlib.pyplot as plt
```

```python
import matlab.engine
eng = matlab.engine.start_matlab("-desktop")
eng.setupVGG(nargout=0)
eng.addpath('C:\Users\zhenr\Dropbox\CMU\Robostats\Final Project', nargout=0)
```

```python
fairwaydir = 'H:/Grass/testset/fairway/'
fairwayfiles = [fairwaydir + f for f in os.listdir(fairwaydir) if isfile(join(fairwaydir, f))]
roughdir = 'H:/Grass/testset/rough/'
roughfiles = [roughdir + f for f in os.listdir(roughdir) if isfile(join(roughdir, f))]

numfairway = len(fairwayfiles)
numrough = len(roughfiles)
#print fairwayfiles
```

```python
clf = [SGDClassifier(loss="hinge", penalty="l2"),
SGDClassifier(loss="hinge", penalty="l2"),
SGDClassifier(loss="hinge", penalty="l2"),
SGDClassifier(loss="hinge", penalty="l2"),
SGDClassifier(loss="hinge", penalty="l2")]
```

```python
#initialize
fairway_i = 0
rough_i = 0

currclass = 'fairway'
#initialize svm
convs = eng.getVGGConvs(fairwayfiles[0], nargout=5)
for i in range(len(clf)):
    clf[i].partial_fit(convs[i], ['fairway'], ['fairway', 'rough'])

mistakes = [0, 0, 0, 0, 0]
total = 0
pred = ''

averageMistakes= [
    [],[],[],[],[]
]
```

```python
#train
while fairway_i < numfairway or rough_i < numrough:
    if fairway_i < numfairway and currclass == 'fairway':
        convs = eng.getVGGConvs(fairwayfiles[fairway_i], nargout=5)
        fairway_i += 1
        total += 1
        
        for i in range(len(clf)):
            pred = clf[i].predict(convs[i])

            if (pred[0] != 'fairway'):
                mistakes[i] += 1
                clf[i].partial_fit(convs[i], ['fairway'])
                
            averageMistakes[i].append(mistakes[i]/float(total))
            

        if (fairway_i+1) % 20 == 0:
            print 'SWITCHING'
            currclass = 'rough'
            
    elif rough_i < numrough and currclass == 'rough':
        convs = eng.getVGGConvs(roughfiles[rough_i], nargout=5)
        rough_i += 1
        total += 1
        
        for i in range(len(clf)):
            pred = clf[i].predict(convs[i])     
            if (pred[0] != 'rough'):
                mistakes[i] += 1
                clf[i].partial_fit(convs[i], ['rough'])
                
            averageMistakes[i].append(mistakes[i]/float(total))

        if (rough_i+1) % 20 == 0:
            print 'SWITCHING'
            currclass = 'fairway'
            
    elif rough_i >= numrough and currclass == 'rough':
        currclass = 'fairway'
    elif fairway_i >= numfairway and currclass == 'fairway':
        currclass = 'rough'
        
    
 
    print("truth: %s, pred: %s" % (currclass, pred))
    print("%d, %d" % (fairway_i, rough_i))
    print total
    print mistakes
        


```

```python
%matplotlib qt5
conv1, = plt.plot(averageMistakes[0], label='conv1')
conv2, = plt.plot(averageMistakes[1], label='conv2')
conv3, = plt.plot(averageMistakes[2], label='conv3')
conv4, = plt.plot(averageMistakes[3], label='conv4')
conv5, = plt.plot(averageMistakes[4], label='conv5')
plt.legend(handles=[conv1, conv2, conv3, conv4, conv5])
plt.xlabel('Image Num')
plt.ylabel('Average Mistakes')
plt.show()
```
